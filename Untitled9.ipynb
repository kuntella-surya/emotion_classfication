```python
import pandas as pd

df = pd.read_csv("/content/tripadvisor_hotel_reviews.csv")
```

```python
import re
import pandas as pd

def preprocess_text(text):
    # handle missing values
    if pd.isna(text):
        return ""

    # 1. lowercase
    text = text.lower()

    # 2. remove HTML tags (if any)
    text = re.sub(r'<.*?>', '', text)

    # 3. remove special characters but KEEP words structure
    text = re.sub(r'[^a-z\s]', '', text)

    # 4. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
df['clean_review'] = df['Review'].apply(preprocess_text)
```

```python
import numpy as np
import nltk
nltk.download('punkt_tab')
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
sentences = df['clean_review'].apply(word_tokenize).tolist()
```

```python
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

```python
def vectorize(sentence):
    words = word_tokenize(sentence)
    vectors = []

    for word in words:
        if word in w2v_model.wv:
            vectors.append(w2v_model.wv[word])

    if len(vectors) == 0:
        return np.zeros(100)

    return np.mean(vectors, axis=0)
```

```python
X = np.array(df['clean_review'].apply(vectorize).tolist())
y = df['Rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Word2Vec Accuracy:", accuracy_score(y_test, y_pred))
```

```python
df['label'] = df['Rating'] - 1
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    df[['clean_review','label']],
    test_size=0.2,
    random_state=42,
    stratify=df['label']
)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained(
    "bert-base-uncased"
)
```

```python
train_encodings = tokenizer(
    train_df['clean_review'].tolist(),
    truncation=True,
    padding=True,
    max_length=128
)

test_encodings = tokenizer(
    test_df['clean_review'].tolist(),
    truncation=True,
    padding=True,
    max_length=128
)
```

```python
import torch

class ReviewDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = ReviewDataset(
    train_encodings,
    train_df['label'].tolist()
)

test_dataset = ReviewDataset(
    test_encodings,
    test_df['label'].tolist()
)
```

```python
from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=5
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    weight_decay=0.01
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)
```

```python
trainer.train()
predictions = trainer.predict(test_dataset)
trainer.evaluate()
```
